# BP fixed points: K=3 assortative

Reproduces TABLE2 from the original 3-ass script using the unified `BPConfig` / `BeliefPropagation` API.
Logs to wandb (project `bp_fixed_point`, group `3ASS_TABLE2`) and saves locally to `results/bp/`.

```python
import sys, os
sys.path.insert(0, '..')  # adjust if you move this notebook

import numpy as np
from src.BP import BPConfig, BeliefPropagation, run_bp
```

## Single run

```python
cfg = BPConfig(
    K=3, d=7, H=3,
    problem_type='assortative',
    m_target=np.array([1/3, 1/3, 1/3]),
    max_iter=10_000_000,
    threshold=1e-15,
    damping=0.01,
    init_type='almost_unif',
    mu_mode='always_zero',
    seed=42,
    log_every=1000,
    use_wandb=True,             # set True once you've done `wandb login`
    wandb_project='bp_fixed_point',
    wandb_group='TEST1',
    save_locally=True,
    save_dir='../results/bp',
)
result = run_bp(cfg, verbose=2)
print(f'phi_RS = {result.phi_RS:.6f}')
print(f's     = {result.s:.6f}')
print(f'm     = {result.m_actual}')
```

## Sweep (one run per (D, H), N_RUNS seeds each)

```python
Ds_Hs = [(11, 4), (11, 5), (11, 6)]
N_RUNS = 3

results = []
for D, H in Ds_Hs:
    for run_id in range(N_RUNS):
        cfg = BPConfig(
            K=3, d=D, H=H,
            problem_type='assortative',
            m_target=np.array([1/3, 1/3, 1/3]),
            max_iter=10_000_000,
            threshold=1e-17,
            damping=0.01,
            init_type='almost_unif',
            mu_mode='always_zero',
            seed=run_id,
            log_every=1000,
            use_wandb=False,
            wandb_project='bp_fixed_point',
            wandb_group='3ASS_TABLE2',
            wandb_name=f'D{D}_H{H}_run{run_id}',
            save_locally=True,
            save_dir='../results/bp',
        )
        res = run_bp(cfg, verbose=1)
        results.append({
            'D': D, 'H': H, 'run': run_id,
            'phi_RS': res.phi_RS, 's': res.s, 'iters': res.iters,
            'converged': res.converged,
        })

import pandas as pd
df = pd.DataFrame(results)
df
```
